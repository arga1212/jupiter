```python
import matplotlib.pyplot as plt

x = [10, 20, 35, 40]
y = [30, 45, 60, 70]

plt.xlabel('Overs')
plt.ylabel('Runs Scored')
plt.title('Over wise runs scored \n Indonesia vs Thailand')
plt.plot(
    x, y, 'r',
    linewidth=4,
    linestyle='dashed',
    marker='+',
    markersize=10,
    markeredgecolor='blue'
)
plt.show()
```

```python
import matplotlib.pyplot as plt

Mobil = ['Honda Civic', 'Toyota Camry', 'Ford Mustang', 'Chevrolet Camaro', 'Tesla Model 3']
KecepatanMaksimum = [200, 190, 250, 245, 261]
plt.bar(Mobil, KecepatanMaksimum, color=['blue', 'red', 'green', 'purple', 'orange'])
plt.xlabel('Mobil')
plt.ylabel('Kecepatan Maksimum (km/jam)')
plt.title('Kecepatan Maksimum Berbagai Mobil')
plt.show()
```

```python
import matplotlib.pyplot as plt
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80]
years=[1, 3, 4, 5, 7, 8, 9, 10, 12, 15, 18, 20, 22]
plt.hist(age, bins=years, color='blue', histtype='bar', rwidth=.8)
plt.xlabel('Emp Age')
plt.ylabel('Number of Emp')
plt.title('KVS')
plt.legend()
plt.show()
```
